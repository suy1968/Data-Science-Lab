Take a string, make a list of strings where each element is the original string without a character. For example, "Jake" => ["ake", "Jke", "Jae", "Jak"]

```python
s= "jake"
str_list= [''.join(s[i] for i in range(len(s)) if i != j) for j in range(len(s)) ]
print(str_list)



```
